```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
def ler_csv(file, separador, encode, zebrado=False):
    return pd.read_csv(file, sep=separador, encoding=encode, squeeze = zebrado)
```

```python
censo = ler_csv("ESCOLAS.csv", '|', 'iso-8859-1', True)
print(censo)
```

```python
saeb = ler_csv("TS_ESCOLA.csv", ',', 'iso-8859-1', True)
print(saeb)
```

```python
censo.rename(columns = {"CO_ENTIDADE":"ID_ESCOLA"}, inplace=True)
print(censo)
```

```python
censo_saeb = pd.merge(censo, saeb, on = "ID_ESCOLA")
print(censo_saeb)
```

```python
censo_saeb.to_csv("censo_saeb_join.csv", sep="|", encoding='iso-8859-1')
```
